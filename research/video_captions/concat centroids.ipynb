```python
import os
from glob import glob
import numpy as np
```

```python
centroids_fp = 'data/centroids'
```

```python
centroids_files = glob(centroids_fp+'/*')
```

```python
centroids_files
```

```python
new = np.load(centroids_files[0])
for file in centroids_files[1:]:
    x = np.load(file)
    new = np.concatenate([new,x])
```

```python
new.shape
```

```python
np.save('data/centroids_prepared/centroids', new)
```
